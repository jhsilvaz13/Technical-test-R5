# Uniqueness
---
La unicidad es una propiedad de los conjuntos que permite identificar de manera única cada uno de los elementos de estos. Para asegurar que un conjunto sea único, se debe cumplir que cada uno de sus elementos sea único, es decir, que no exista otro elemento igual dentro del conjunto. En este caso se debe verificar que solo existe una canción registrada, debido a que cada registro representa una canción de Taylor Swift, por lo que no puede existir otra canción con el mismo identificador. Primero se realiza un conteo básico donde se verifica con la librería de Pandas que no existan registros duplicados, teniendo en cuenta que los registros duplicados son registros cuyos campos son exactamente iguales.

```python
df_duplicados=df[df.duplicated()]
df_duplicados
```

Como vemos existe un total de 18 registros duplicados, en el proceso de limpieza se recomienda eliminar estos registros para evitar inconsistencias en los datos. 

## Verificar llave primaria para canciones

Despues de haber contado los registros duplicados, se procede a identificar la llave primaria del dataset, en este dataset existe un total de 4 columnas que representan identificadores, estas columnas son:

- artist_id
- album_id
