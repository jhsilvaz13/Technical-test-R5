# Validity
---
Esta dimensión mide la cantidad de datos que coincide con el formato requerido para cualquier regla de negocio. El formato generalmente incluye metadatos, como tipos de datos válidos, rangos, patrones y más(ibm). Para calcular un porcentaje de validez de datos se llevara a cabo un proceso dividido en 4 etapas:
1. Consistencia en los tipos de datos
2. Consistencia en las unidades
3. Consistencia en los rangos
4. Consistencia en las categorías
5. Consistencia en las cadenas de caracteres

Para esto se recuerda el significado de cada una de las columnas que se encuentran en el dataset, esta documentación se encuentra en el siguiente [Spotify Web API](https://developer.spotify.com/documentation/web-api).
* **disc_number:** The disc number (usually 1 unless the album consists of more than one disc).
* **duration_ms:** The track length in milliseconds.
* **explicit:** Whether or not the track has explicit lyrics ( true = yes it does; false = no it does not OR unknown).
* **track_number:** The number of the track. If an album has several discs, the track number is the number on the specified disc.
* **track_popularity:** The popularity of the track. The value will be between 0 and 100, with 100 being the most popular. The popularity of a track is a value between 0 and 100, with 100 being the most popular. The popularity is calculated by algorithm and is based, in the most part, on the total number of plays the track has had and how recent those plays are. Generally speaking, songs that are being played a lot now will have a higher popularity than songs that were played a lot in the past. Duplicate tracks (e.g. the same track from a single and an album) are rated independently. Artist and album popularity is derived mathematically from track popularity. Note: the popularity value may lag actual popularity by a few days: the value is not updated in real time.
* **track_id:** The Spotify ID for the track.
* **track_name:** The name of the track.
* **audio_features.danceability:** Danceability describes how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity. A value of 0.0 is least danceable and 1.0 is most danceable.
* **audio_features.energy:** Energy is a measure from 0.0 to 1.0 and represents a perceptual measure of intensity and activity. Typically, energetic tracks feel fast, loud, and noisy. For example, death metal has high energy, while a Bach prelude scores low on the scale. Perceptual features contributing to this attribute include dynamic range, perceived loudness, timbre, onset rate, and general entropy.
* **audio_features.key:** The key the track is in. Integers map to pitches using standard Pitch Class notation. E.g. 0 = C, 1 = C♯/D♭, 2 = D, and so on. If no key was detected, the value is -1. The range is -1 to 11.
* **audio_features.loudness:** The overall loudness of a track in decibels (dB). Loudness values are averaged across the entire track and are useful for comparing relative loudness of tracks. Loudness is the quality of a sound that is the primary psychological correlate of physical strength (amplitude). Values typically range between -60 and 0 db
* **audio_features.mode:** Mode indicates the modality (major or minor) of a track, the type of scale from which its melodic content is derived. Major is represented by 1 and minor is 0.
* **audio_features.speechiness:** Speechiness detects the presence of spoken words in a track. The more exclusively speech-like the recording (e.g. talk show, audio book, poetry), the closer to 1.0 the attribute value. Values above 0.66 describe tracks that are probably made entirely of spoken words. Values between 0.33 and 0.66 describe tracks that may contain both music and speech, either in sections or layered, including such cases as rap music. Values below 0.33 most likely represent music and other non-speech-like tracks.
* **audio_features.acousticness:** A confidence measure from 0.0 to 1.0 of whether the track is acoustic. 1.0 represents high confidence the track is acoustic.
* **audio_features.instrumentalness:** Predicts whether a track contains no vocals. "Ooh" and "aah" sounds are treated as instrumental in this context. Rap or spoken word tracks are clearly "vocal". The closer the instrumentalness value is to 1.0, the greater likelihood the track contains no vocal content. Values above 0.5 are intended to represent instrumental tracks, but confidence is higher as the value approaches 1.0.
* **audio_features.liveness:** Detects the presence of an audience in the recording. Higher liveness values represent an increased probability that the track was performed live. A value above 0.8 provides strong likelihood that the track is live.
* **audio_features.valence:** A measure from 0.0 to 1.0 describing the musical positiveness conveyed by a track. Tracks with high valence sound more positive (e.g. happy, cheerful, euphoric), while tracks with low valence sound more negative (e.g. sad, depressed, angry).
* **audio_features.tempo:** The overall estimated tempo of a track in beats per minute (BPM). In musical terminology, tempo is the speed or pace of a given piece and derives directly from the average beat duration.
* **audio_features.id:** 
* **audio_features.time_signature:** An estimated time signature. The time signature (meter) is a notational convention to specify how many beats are in each bar (or measure). The time signature ranges from 3 to 7 indicating time signatures of "3/4", to "7/4".
* **artist_id:** The Spotify ID for the artist.

Al identificar estas inconsistencias en los tipos de datos se calculara un porcentaje parcial de consistencia de tipos de datos para cada una de las columnas que tienen inconsistencias en los tipos de datos. Para esto se utiliza la siguiente fórmula:

$$
\text{Porcentaje de consistencia de tipos de datos} = \frac{\text{Número de registros que coinciden con el tipo de dato deseado}}{\text{Número total de registros}}
$$

Esta se aplica para cada una de las columnas.


```python
df_types=df.copy()
df_types
```

```python
#For numeric values we try to convert calues to numeric and if is not possible this will return a NaN value,after that we count the number of NaN values
numeric_columns=['disc_number',
                 'duration_ms',
                 'track_number',
                 'track_popularity',
                 'audio_features.danceability',
                 'audio_features.energy',
                 'audio_features.key',
                 'audio_features.loudness',
                 'audio_features.mode',
                 'audio_features.speechiness',
                 'audio_features.acousticness',
                 'audio_features.instrumentalness',
                 'audio_features.liveness',
                 'audio_features.valence',
                 'audio_features.tempo',
                 'audio_features.time_signature',
                 'artist_popularity',
                 'album_total_tracks']
for column in numeric_columns:
    column_temp=df[column]# save the column in a temp variable
    column_temp=column_temp[column_temp.notna()]# remove the NaN values if we don't do this at the end we count thus values as types inconsistencies
    column_temp=pd.to_numeric(column_temp, errors='coerce')
    print('The column', column, 'has', column_temp.isna().sum(), 'values that are not numeric')
    
```

```python
#For boolean values we try to convert values to boolean and the values that are not possible to convert will return a NaN value, after that we count the number of NaN values
#also we check before the conversion if there is any value that is null because this represents an ausence of value and not a type inconsistency
boolean_columns=['explicit']
column_temp=df[boolean_columns[0]]# save the column in a temp variable
column_temp.replace(
    {'True': True,
     'False': False,
    }
    , inplace=True)
column_temp=pd.to_numeric(column_temp, errors='coerce')
column_temp=column_temp.astype('boolean', errors='raise')
print('The column', boolean_columns[0], 'has', column_temp.isna().sum(), 'values that are not boolean')
```

```python
# For datetimes values we try to convert values to datetime and the values that are not possible to convert will return a NaN value, after that we count the number of NaN values
datetime_columns=['album_release_date']
column_temp=df[datetime_columns[0]]# save the column in a temp variable
column_temp=pd.to_datetime(column_temp, errors='coerce')
print('The column', datetime_columns[0], 'has', column_temp.isna().sum(), 'values that are not datetime')
```

Ahora se resume esta informacion en una tabla que nos muestre la cantidad de valores correctamente tipados y el porcentaje que representan del total de registros por culumnas. Para las cadenas de caracteres se conoce que el porcentaje es $100\%$ pues cualquier cadena incluso si representa un número u otrop tipo de dato es correcta.

```python
numeric_columns=['disc_number',
                 'duration_ms',
                 'track_number',
                 'track_popularity',
                 'audio_features.danceability',
                 'audio_features.energy',
                 'audio_features.key',
                 'audio_features.loudness',
                 'audio_features.mode',
                 'audio_features.speechiness',
                 'audio_features.acousticness',
                 'audio_features.instrumentalness',
                 'audio_features.liveness',
                 'audio_features.valence',
                 'audio_features.tempo',
                 'audio_features.time_signature',
                 'artist_popularity',
                 'album_total_tracks']
boolean_columns=['explicit']
datetime_columns=['album_release_date']
object_columns=['track_id',
                'track_name',
                'audio_features.id',
                'artist_id',
                'artist_name',
                'album_id',
                'album_name',]
df_types=pd.DataFrame(columns=['Column_name', 'Type', 'Correct types','Percentage of correct types'])

for column in df.columns:
    if column in object_columns:
        df_types.loc[len(df_types)]={
            'Column_name': column,
            'Type': 'object',
            'Correct types': len(df[column]),
            'Percentage of correct types': 100
        }
    elif column in numeric_columns:
        column_temp=df[column]
        column_temp=column_temp[column_temp.notna()]
        column_temp=pd.to_numeric(column_temp, errors='coerce')
        df_types.loc[len(df_types)]={
            'Column_name': column,
            'Type': 'numeric',
            'Correct types':len(df[column])-column_temp.isna().sum(),
            'Percentage of correct types': (len(df[column])-column_temp.isna().sum())/len(df[column])*100
        }
    elif column in boolean_columns:
        column_temp=df[column]# save the column in a temp variable
        column_temp.replace(
            {'True': True,
            'False': False,
            }
            , inplace=True)
        column_temp=pd.to_numeric(column_temp, errors='coerce')
        column_temp=column_temp.astype('boolean', errors='raise')
        df_types.loc[len(df_types)]={
            'Column_name': column,
            'Type': 'boolean',
            'Correct types':len(df[column])-column_temp.isna().sum(),
            'Percentage of correct types': (len(df[column])-column_temp.isna().sum())/len(df[column])*100
        }
    elif column in datetime_columns:
        datetime_columns=['album_release_date']
        column_temp=df[datetime_columns[0]]# save the column in a temp variable
        column_temp=pd.to_datetime(column_temp, errors='coerce')
        df_types.loc[len(df_types)]={
            'Column_name': column,
            'Type': 'datetime',
            'Correct types':len(df[column])-column_temp.isna().sum(),
            'Percentage of correct types': (len(df[column])-column_temp.isna().sum())/len(df[column])*100
        }
df_types
```

Para continuar evaluando la dimensión de **Validez** se remplazan los valores que no coinciden con el tipo de dato.

```python
df['explicit'].replace({'Si': "True", 'No': "False"}, inplace=True)
df['explicit']=df['explicit'].astype(bool)
df['album_total_tracks']=df['album_total_tracks'].replace({'Thirteen': 13}).astype(int)
df['audio_features.instrumentalness']=df['audio_features.instrumentalness'].replace({'7.28x-06':7.20e-06}).astype(float)
```

El porcentaje de consistencia en rangos se calcula con la siguiente fórmula:

***Nota***

```python
duration_ms_range=[0, ]
track_popularity_range=[0, 100]
audio_features_range=[0, 1]
df
```

```python
# Count for the columns that previously we identified that has values out of range, how much values are out of range
columns_out_of_range=['duration_ms',
                      'track_popularity',
                      'audio_features.acousticness']

df_range=pd.DataFrame(columns=['Column_name', 'Values in correct range', 'Percentage of values in correct range'])

for column in df.columns:
    column_temp=df[column]
    if column in columns_out_of_range:
        if column=='duration_ms':
            df_range.loc[len(df_range)]={
                'Column_name': column,
                'Values in correct range': len(column_temp)-len(column_temp[(column_temp<duration_ms_range[0]) | (column_temp>duration_ms_range[1])]),
                'Percentage of values in correct range': (len(column_temp)-len(column_temp[(column_temp<duration_ms_range[0]) | (column_temp>duration_ms_range[1])]))/len(column_temp)*100
            }
        elif column=='track_popularity':
            df_range.loc[len(df_range)]={
                'Column_name': column,
                'Values in correct range': len(column_temp)-len(column_temp[(column_temp<track_popularity_range[0]) | (column_temp>track_popularity_range[1])]),
                'Percentage of values in correct range': (len(column_temp)-len(column_temp[(column_temp<track_popularity_range[0]) | (column_temp>track_popularity_range[1])]))/len(column_temp)*100
            }
        elif column=='audio_features.acousticness':
            df_range.loc[len(df_range)]={
                'Column_name': column,
                'Values in correct range': len(column_temp)-len(column_temp[(column_temp<audio_features_range[0]) | (column_temp>audio_features_range[1])]),
                'Percentage of values in correct range': (len(column_temp)-len(column_temp[(column_temp<audio_features_range[0]) | (column_temp>audio_features_range[1])]))/len(column_temp)*100
            }            
    else:
        df_range.loc[len(df_range)]={
            'Column_name': column,
            'Values in correct range': len(df[column]),
            'Percentage of values in correct range': 100
        }
df_range
        
```
